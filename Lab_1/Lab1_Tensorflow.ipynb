```python
import tensorflow as tf

# Tạo một tensor có kích thước [2, 3] với các phần tử được gán 0
zero_tensor = tf.zeros([2, 3])

# Hiển thị tensor
print(zero_tensor)
```

```python
X = tf.constant([[1, 2, 3], [4, 5, 6]])

zero_like_X_tensor = tf.zeros_like(X)

print("Tensor X:")
print(X)
print("\nTensor zero_like_X_tensor:")
print(zero_like_X_tensor)
```

```python
# Tạo một tensor có kích thước [3, 2] với các phần tử được gán 5
five_tensor = tf.fill([3, 2], 5)

# Hiển thị tensor
print(five_tensor)
```

```python
# Xác định các số chẵn trong khoảng [5, 10]
even_numbers_in_range = tf.constant([6, 8, 10])

# Tính toán số lần lặp lại cần thiết để có ít nhất 50 phần tử
num_repetitions = tf.cast(tf.math.ceil(50 / tf.cast(tf.size(even_numbers_in_range), dtype=tf.float32)), dtype=tf.int32)

# Lặp lại tensor và cắt để có chính xác 50 phần tử
tensor_50_even = tf.tile(even_numbers_in_range, [num_repetitions])[:50]

print("Tensor 1-chiều với 50 phần tử chẵn:")
print(tensor_50_even)
print("Kích thước tensor:", tensor_50_even.shape)
```

```python
# Tạo một tensor có kích thước [3, 2] với giá trị ngẫu nhiên theo phân phối normal
# mean=0, standard deviation = 2
random_normal_tensor = tf.random.normal(shape=[3, 2], mean=0.0, stddev=2.0)

# Hiển thị tensor
print(random_normal_tensor)
```

```python
X = tf.constant([[1, 2], [3, 4], [5, 6]])

shuffled_X = tf.random.shuffle(X)

print("Tensor X ban đầu:")
print(X)
print("\nTensor X sau khi xáo trộn:")
print(shuffled_X)
```

```python
# Định nghĩa hai tensor x và y
x = tf.constant([1, 2, 3])
y = tf.constant([4, 5, 6])

# Tính tổng của hai tensor
sum_tensor = tf.add(x, y) # Hoặc đơn giản là x + y

# Hiển thị kết quả
print("Tensor x:", x)
print("Tensor y:", y)
print("Tổng của x và y:", sum_tensor)
```

```python
# Tạo một tensor ngẫu nhiên kích thước (10, 10, 3)
original_tensor = tf.random.uniform(shape=(10, 10, 3), minval=0, maxval=255, dtype=tf.int32)

# Cắt ngẫu nhiên tensor thành kích thước (5, 5, 3)
cropped_tensor = tf.image.random_crop(original_tensor, size=[5, 5, 3])

print("Tensor gốc (kích thước", original_tensor.shape, "):")
print(original_tensor)
print("\nTensor đã cắt (kích thước", cropped_tensor.shape, "):")
print(cropped_tensor)
```
