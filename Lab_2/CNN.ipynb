```python
# ============================================================
# CODE CNN MNIST
# ============================================================

# 1. Cài đặt và import thư viện
!pip install -q tensorflow matplotlib seaborn scikit-learn

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import mnist
from sklearn.metrics import confusion_matrix

print("✅ Import thư viện thành công!")
print(f"TensorFlow version: {keras.__version__}")

(X_train, y_train), (X_test, y_test) = mnist.load_data()

print(f"\n📊 Thông tin dữ liệu:")
print(f"  - X_train shape: {X_train.shape}")
print(f"  - y_train shape: {y_train.shape}")
print(f"  - X_test shape: {X_test.shape}")
print(f"  - y_test shape: {y_test.shape}")

# 3. Tiền xử lý dữ liệu
# Chuẩn hóa pixel về [0, 1]
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# Thêm chiều channel (28, 28) -> (28, 28, 1)
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

# One-hot encoding cho nhãn
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

print(f"\n✅ Sau tiền xử lý:")
print(f"  - X_train shape: {X_train.shape}")
print(f"  - y_train shape: {y_train.shape}")

# 4. Trực quan hóa 10 ảnh đầu tiên
fig, axes = plt.subplots(2, 5, figsize=(12, 6))
axes = axes.ravel()
for i in range(10):
    axes[i].imshow(X_train[i].reshape(28, 28), cmap='gray')
    axes[i].set_title(f'Label: {np.argmax(y_train[i])}')
    axes[i].axis('off')
plt.suptitle('10 ẢNH ĐẦU TIÊN - MNIST', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# 5. Xây dựng mô hình CNN
model = Sequential()

# Input layer
model.add(Input(shape=(28, 28, 1)))

# Convolutional block 1
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Convolutional block 2
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Flatten và Fully Connected
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))  # Chống overfitting
model.add(Dense(10, activation='softmax'))

# Hiển thị kiến trúc
model.summary()

# 6. Biên dịch mô hình
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\n✅ Biên dịch mô hình thành công!")

# 7. Huấn luyện mô hình
history = model.fit(
    X_train, y_train,
    batch_size=128,
    epochs=10,
    validation_split=0.1,
    verbose=1
)

print("\n✅ Huấn luyện hoàn tất!")

# 8. Đánh giá trên tập test
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
print(f"\n{'='*50}")
print(f"📊 KẾT QUẢ TRÊN TẬP TEST:")
print(f"  - Test Loss: {test_loss:.4f}")
print(f"  - Test Accuracy: {test_acc:.4f}")
print(f"  - Tỷ lệ đúng: {test_acc*100:.2f}%")
print(f"{'='*50}")

# 9. Vẽ biểu đồ Accuracy và Loss
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Accuracy plot
axes[0].plot(history.history['accuracy'], 'b-', label='Train', linewidth=2)
axes[0].plot(history.history['val_accuracy'], 'r-', label='Validation', linewidth=2)
axes[0].set_title('Model Accuracy', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend(loc='lower right')
axes[0].grid(True, alpha=0.3)

# Loss plot
axes[1].plot(history.history['loss'], 'b-', label='Train', linewidth=2)
axes[1].plot(history.history['val_loss'], 'r-', label='Validation', linewidth=2)
axes[1].set_title('Model Loss', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend(loc='upper right')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 10. Dự đoán và hiển thị kết quả
y_pred = model.predict(X_test[:16])
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test[:16], axis=1)

fig, axes = plt.subplots(4, 4, figsize=(12, 12))
axes = axes.ravel()

for i in range(16):
    axes[i].imshow(X_test[i].reshape(28, 28), cmap='gray')
    color = 'green' if y_pred_classes[i] == y_true_classes[i] else 'red'
    axes[i].set_title(f'True: {y_true_classes[i]}\nPred: {y_pred_classes[i]}',
                      color=color, fontsize=10)
    axes[i].axis('off')

plt.suptitle('DỰ ĐOÁN MNIST (Xanh: Đúng, Đỏ: Sai)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# 11. Ma trận nhầm lẫn
y_pred_all = model.predict(X_test)
y_pred_classes_all = np.argmax(y_pred_all, axis=1)
y_true_all = np.argmax(y_test, axis=1)

cm = confusion_matrix(y_true_all, y_pred_classes_all)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=range(10), yticklabels=range(10))
plt.title('Confusion Matrix - MNIST', fontsize=14, fontweight='bold')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# 12. Lưu model
model.save('/content/drive/MyDrive/THDL/mnist_cnn_model.h5')
print("\n✅ Đã lưu model thành công: mnist_cnn_model.h5")
```

```python
# ============================================================
# BÀI TẬP 1: CIFAR-10 - NHẬN DẠNG 10 LOẠI ĐỐI TƯỢNG
# ============================================================

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, Input
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import cifar10
from sklearn.metrics import confusion_matrix, classification_report
import warnings
warnings.filterwarnings('ignore')

print("="*60)
print("BÀI TẬP 1: CIFAR-10 CLASSIFICATION")
print("="*60)

# 1. Tải dữ liệu CIFAR-10
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# Tên các lớp
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

print(f"\n📊 Thông tin dữ liệu:")
print(f"  - X_train shape: {X_train.shape}")
print(f"  - y_train shape: {y_train.shape}")
print(f"  - X_test shape: {X_test.shape}")

# 2. Tiền xử lý dữ liệu
# Chuẩn hóa pixel về [0, 1]
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# One-hot encoding cho nhãn
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

print(f"\n✅ Sau tiền xử lý:")
print(f"  - X_train shape: {X_train.shape}")
print(f"  - y_train shape: {y_train.shape}")

# 3. Trực quan hóa 16 ảnh đầu tiên
plt.figure(figsize=(12, 10))
for i in range(16):
    plt.subplot(4, 4, i+1)
    plt.imshow(X_train[i])
    plt.title(class_names[np.argmax(y_train[i])], fontsize=10)
    plt.axis('off')
plt.suptitle('MẪU ẢNH CIFAR-10', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/THDL/cifar10_samples.png', dpi=150)
plt.show()

# 4. Xây dựng mô hình CNN cho CIFAR-10 (kiến trúc sâu hơn)
model_cifar = Sequential()

# Input
model_cifar.add(Input(shape=(32, 32, 3)))

# Block 1
model_cifar.add(Conv2D(32, (3, 3), activation='relu', padding='same'))
model_cifar.add(BatchNormalization())
model_cifar.add(Conv2D(32, (3, 3), activation='relu'))
model_cifar.add(MaxPooling2D((2, 2)))
model_cifar.add(Dropout(0.25))

# Block 2
model_cifar.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model_cifar.add(BatchNormalization())
model_cifar.add(Conv2D(64, (3, 3), activation='relu'))
model_cifar.add(MaxPooling2D((2, 2)))
model_cifar.add(Dropout(0.25))

# Block 3
model_cifar.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model_cifar.add(BatchNormalization())
model_cifar.add(Conv2D(128, (3, 3), activation='relu'))
model_cifar.add(MaxPooling2D((2, 2)))
model_cifar.add(Dropout(0.25))

# Fully Connected
model_cifar.add(Flatten())
model_cifar.add(Dense(512, activation='relu'))
model_cifar.add(BatchNormalization())
model_cifar.add(Dropout(0.5))
model_cifar.add(Dense(10, activation='softmax'))

# Hiển thị kiến trúc
model_cifar.summary()

# 5. Biên dịch mô hình
model_cifar.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# 6. Callbacks
callbacks = [
    keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True),
    keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=3, verbose=1)
]

# 7. Huấn luyện
print("\n🚀 Bắt đầu huấn luyện CIFAR-10...")
history_cifar = model_cifar.fit(
    X_train, y_train,
    batch_size=64,
    epochs=30,
    validation_split=0.1,
    callbacks=callbacks,
    verbose=1
)

# 8. Đánh giá trên tập test
test_loss, test_acc = model_cifar.evaluate(X_test, y_test, verbose=0)
print(f"\n{'='*50}")
print(f"📊 KẾT QUẢ CIFAR-10 TRÊN TẬP TEST:")
print(f"  - Test Loss: {test_loss:.4f}")
print(f"  - Test Accuracy: {test_acc:.4f}")
print(f"  - Tỷ lệ đúng: {test_acc*100:.2f}%")
print(f"{'='*50}")

# 9. Vẽ biểu đồ kết quả
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(history_cifar.history['accuracy'], 'b-', label='Train', linewidth=2)
axes[0].plot(history_cifar.history['val_accuracy'], 'r-', label='Validation', linewidth=2)
axes[0].set_title('CIFAR-10 Accuracy', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].plot(history_cifar.history['loss'], 'b-', label='Train', linewidth=2)
axes[1].plot(history_cifar.history['val_loss'], 'r-', label='Validation', linewidth=2)
axes[1].set_title('CIFAR-10 Loss', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('/content/drive/MyDrive/THDL/cifar10_results.png', dpi=150)
plt.show()

# 10. Dự đoán và hiển thị kết quả
y_pred = model_cifar.predict(X_test[:16])
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test[:16], axis=1)

fig, axes = plt.subplots(4, 4, figsize=(12, 12))
axes = axes.ravel()

for i in range(16):
    axes[i].imshow(X_test[i])
    color = 'green' if y_pred_classes[i] == y_true_classes[i] else 'red'
    axes[i].set_title(f'True: {class_names[y_true_classes[i]]}\nPred: {class_names[y_pred_classes[i]]}',
                      color=color, fontsize=9)
    axes[i].axis('off')

plt.suptitle('DỰ ĐOÁN CIFAR-10 (Xanh: Đúng, Đỏ: Sai)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/THDL/cifar10_predictions.png', dpi=150)
plt.show()

# 11. Classification Report
y_pred_all = model_cifar.predict(X_test)
y_pred_classes_all = np.argmax(y_pred_all, axis=1)
y_true_all = np.argmax(y_test, axis=1)

print("\n📋 CLASSIFICATION REPORT:")
print(classification_report(y_true_all, y_pred_classes_all, target_names=class_names))

# 12. Confusion Matrix
cm = confusion_matrix(y_true_all, y_pred_classes_all)
plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title('CIFAR-10 Confusion Matrix', fontsize=14, fontweight='bold')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/THDL/cifar10_confusion_matrix.png', dpi=150)
plt.show()

# 13. Lưu model
model_cifar.save('/content/drive/MyDrive/THDL/cifar10_model.h5')
print("\n✅ Đã lưu model: cifar10_model.h5")
```

```python
# ============================================================
# BÀI TẬP 3: FASHION-MNIST - NHẬN DẠNG SẢN PHẨM THỜI TRANG
# ============================================================

from tensorflow.keras.datasets import fashion_mnist

print("="*60)
print("BÀI TẬP 3: FASHION-MNIST CLASSIFICATION")
print("="*60)

# 1. Tải dữ liệu
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

print(f"\n📊 Thông tin dữ liệu:")
print(f"  - X_train shape: {X_train.shape}")
print(f"  - X_test shape: {X_test.shape}")

# 2. Tiền xử lý
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

print(f"\n✅ Sau tiền xử lý:")
print(f"  - X_train shape: {X_train.shape}")

# 3. Trực quan hóa
plt.figure(figsize=(12, 8))
for i in range(16):
    plt.subplot(4, 4, i+1)
    plt.imshow(X_train[i].reshape(28, 28), cmap='gray')
    plt.title(class_names[np.argmax(y_train[i])], fontsize=9)
    plt.axis('off')
plt.suptitle('FASHION-MNIST SAMPLE IMAGES', fontsize=14, fontweight='bold')
plt.tight_layout()
# plt.savefig('fashion_mnist_samples.png', dpi=150)
plt.show()

# 4. Xây dựng mô hình CNN
model_fashion = Sequential()

model_fashion.add(Input(shape=(28, 28, 1)))
model_fashion.add(Conv2D(32, (3, 3), activation='relu'))
model_fashion.add(MaxPooling2D((2, 2)))
model_fashion.add(Conv2D(64, (3, 3), activation='relu'))
model_fashion.add(MaxPooling2D((2, 2)))
model_fashion.add(Conv2D(128, (3, 3), activation='relu'))
model_fashion.add(MaxPooling2D((2, 2)))
model_fashion.add(Flatten())
model_fashion.add(Dense(128, activation='relu'))
model_fashion.add(Dropout(0.5))
model_fashion.add(Dense(10, activation='softmax'))

model_fashion.summary()

# 5. Biên dịch và huấn luyện
model_fashion.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\n🚀 Bắt đầu huấn luyện FASHION-MNIST...")
history_fashion = model_fashion.fit(
    X_train, y_train,
    batch_size=128,
    epochs=15,
    validation_split=0.1,
    verbose=1
)

# 6. Đánh giá
test_loss, test_acc = model_fashion.evaluate(X_test, y_test, verbose=0)
print(f"\n{'='*50}")
print(f"📊 KẾT QUẢ FASHION-MNIST:")
print(f"  - Test Loss: {test_loss:.4f}")
print(f"  - Test Accuracy: {test_acc:.4f}")
print(f"  - Tỷ lệ đúng: {test_acc*100:.2f}%")
print(f"{'='*50}")

# 7. Vẽ biểu đồ
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(history_fashion.history['accuracy'], 'b-', label='Train', linewidth=2)
axes[0].plot(history_fashion.history['val_accuracy'], 'r-', label='Validation', linewidth=2)
axes[0].set_title('Fashion-MNIST Accuracy', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].plot(history_fashion.history['loss'], 'b-', label='Train', linewidth=2)
axes[1].plot(history_fashion.history['val_loss'], 'r-', label='Validation', linewidth=2)
axes[1].set_title('Fashion-MNIST Loss', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
# plt.savefig('fashion_mnist_results.png', dpi=150)
plt.show()

# 8. Dự đoán mẫu
y_pred = model_fashion.predict(X_test[:16])
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test[:16], axis=1)

fig, axes = plt.subplots(4, 4, figsize=(12, 12))
axes = axes.ravel()

for i in range(16):
    axes[i].imshow(X_test[i].reshape(28, 28), cmap='gray')
    color = 'green' if y_pred_classes[i] == y_true_classes[i] else 'red'
    axes[i].set_title(f'True: {class_names[y_true_classes[i]]}\nPred: {class_names[y_pred_classes[i]]}',
                      color=color, fontsize=9)
    axes[i].axis('off')

plt.suptitle('DỰ ĐOÁN FASHION-MNIST', fontsize=14, fontweight='bold')
plt.tight_layout()
# plt.savefig('fashion_mnist_predictions.png', dpi=150)
plt.show()

# 9. Lưu model
# model_fashion.save('fashion_mnist_model.h5')
# print("\n✅ Đã lưu model: fashion_mnist_model.h5")
```
